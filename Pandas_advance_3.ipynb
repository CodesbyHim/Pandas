```python
# merge and join

# inner join >>> common part , intersection in venn diagram
# left join >>> data in A part only
# right join >>> data in B part only
# Cross join >>> complete data, data from both
# left outer join >>> data from A exclude the common data
# right outer join >>> data from B excluding the common data
```

```python
# merge
```

```python
df1 = pd.DataFrame({'key1': [1,2,4,5,6],'key2':[4,5,6,7,8],'key3':[3,4,5,6,6]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'key1':[1,2,45,6,67],'key4':[56,5,6,7,8],'key5':[3,56,5,6,6]})
```

```python
df2
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2, how = 'left')
```

```python
pd.merge(df1, df2, how = 'right')
```

```python
pd.merge(df1, df2, how = 'outer')  # all the keys from both the dataframe
```

```python
df1
```

```python
pd.merge(df1, df2, how = 'cross')
```

```python
pd.merge(df1, df2, how = 'left', left_on = "key2", right_on = "key4")
```

```python
df1 = pd.DataFrame({'key1': [1,2,4,5,6],
                    'key2':[4,5,6,7,8],
                    'key3':[3,4,5,6,6]
                    },
                   index = ['a', 'b', 'c','d','e'])
```

```python
df2 = pd.DataFrame({'key6':[1,2,45,6,67],
                    'key4':[56,5,6,7,8],
                    'key5':[3,56,5,6,6]
                    },
                   index = ['a','b','h','i','j'])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2, how = "inner") #Join columns of another DataFrame.
```

```python
df1.join(df2, how = 'left')
```

```python
df1.join(df2, how = 'right')
```

```python
df1.join(df2, how = 'outer')
```

```python
df1.join(df2, how = 'cross')
```
