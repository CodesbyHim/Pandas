```python
df2 = pd.DataFrame({"a": [1,2,3,4,5,6,7,8,9]})
df2
```

```python
df2.mean()
```

```python
df2.rolling(window = 2).mean()
# Provide rolling window calculations.
# 1+2 / 2 then 2+3 / 2 continues till last element.
# when we give window = 3 then,
#  1+2+3 / 3 then 2+3+4 / 3 continues till last element.
```

```python
df3 = df2.rolling(window = 3).mean()
```

```python
df3.dropna(inplace=True)
```

```python
df3
```

```python
df3['a_int'] = df3['a'].astype(int)
```

```python
df3
```
